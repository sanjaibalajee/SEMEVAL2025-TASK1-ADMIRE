```python
import pandas as pd
train_df = pd.read_csv('./notebooks/train_with_COT.csv')
train_df.head()
```

```python
prompt = """Below is an instruction that describes a task, paired with an input that provides further context.
Write a response that appropriately completes the request.
Before answering, think carefully about the question and create a step-by-step chain of thoughts to ensure a logical and accurate response.

### Instruction:
You are a linguistic expert specializing in idiomatic expressions, figurative language, and semantic analysis.
Your task is to analyze the given sentence and determine whether the specified phrase is being used in a figurative, idiomatic sense or in its standard, literal meaning.

An **idiomatic phrase** conveys a non-literal meaning that is culturally understood and cannot be directly inferred from the individual words. These expressions often derive from historical, cultural, or metaphorical origins and require familiarity with the language to be correctly interpreted.

A **literal phrase** retains its dictionary definition and can be understood directly from its individual words without requiring additional cultural knowledge.

To determine this:
1. Identify the dictionary meaning of the phrase.
2. Examine its function within the sentence.
3. Assess whether its meaning extends beyond the literal interpretation of the words.
4. If the phrase conveys a widely recognized figurative meaning, classify it as idiomatic.
5. If the phrase maintains its direct, dictionary meaning in context, classify it as literal.

Use logical reasoning to arrive at the most accurate classification.

### Sentence:
{}

### Phrase to Analyze:
'{}'

### Response:
<think>
{}
</think>

The correct answer is: **"{}"**"""

EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    sentences = examples["sentence"]
    compounds = examples["compound"]
    cots = examples["COT"]
    labels = examples["sentence_type"]  # "Yes" or "No" for idiomatic/literal
    texts = []

    for sentence, compound, cot, label in zip(sentences, compounds, cots, labels):
        text = prompt.format(sentence, compound, cot, label) + EOS_TOKEN
        texts.append(text)

    return {
        "text": texts,
    }
```

```python
from datasets import Dataset

# Convert DataFrame to Hugging Face Dataset
train_dataset = Dataset.from_pandas(train_df)

# Apply mapping function
dataset = train_dataset.map(formatting_prompts_func, batched=True)

# Check output
print(dataset["text"][0])
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",  # True or "unsloth" for very long context
    random_state=3407,
    use_rslora=False,
    loftq_config=None,
)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    dataset_num_proc=2,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        # Use num_train_epochs = 1, warmup_ratio for full training runs!
        warmup_steps=5,
        max_steps=60,
        learning_rate=2e-4,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=10,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
    ),
)
```

```python
trainer_stats = trainer.train()
```

```python
def generate_inference_prompt(sentence, phrase):
    prompt = f"""
Below is an instruction that describes a task, paired with an input that provides further context.
Write a concise response that completes the request. Include only a brief explanation of your reasoning if needed, and then provide your final decision.

### Instruction:
You are a linguistic expert specializing in idiomatic expressions, figurative language, and semantic analysis.
Your task is to analyze the given sentence and determine whether the specified phrase is being used in a figurative/idiomatic sense or in its standard/literal meaning.
In your reasoning, please:
- Focus on the simplest and most direct interpretation.
- Analyze only the words immediately surrounding the phrase for clear clues about its intended meaning.
- **If the phrase is being used literally, treat it as its basic meaning.** For instance, consider "black sheep" simply as "a sheep that is black."
- Keep your explanation short (no more than two sentences) and avoid overcomplicating your analysis.
- Conclude with one word: either "literal" or "figurative/idiomatic".

### Sentence:
{sentence}

### Phrase to Analyze:
{phrase}

### Response:
"""
    return prompt.strip()



# Example usage
sentence = "It represented them having signed away their birthright so that they can join their fellow politicians on the gravy train"
phrase = "gravy train"

inference_prompt = generate_inference_prompt(sentence, phrase)
print(inference_prompt)

# Run inference
FastLanguageModel.for_inference(model)  # Ensure Unsloth optimizes for inference
inputs = tokenizer([inference_prompt], return_tensors="pt").to("cuda")  # Use generated prompt

outputs = model.generate(
    input_ids=inputs.input_ids,
    attention_mask=inputs.attention_mask,
    max_new_tokens=1200,
    use_cache=True,
)

# Decode & Extract response properly
response = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
final_response = response.split("### Response:")[1].strip() if "### Response:" in response else response.strip()

# Print the cleaned response
print(final_response)
```
